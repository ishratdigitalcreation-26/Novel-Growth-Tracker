```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from datetime import date

file_name = 'my_progress.csv'

# 1. Purana data load karen ya naye file banayen
if os.path.exists(file_name):
    df = pd.read_csv(file_name)
else:
    df = pd.DataFrame(columns=['Date', 'Views'])

# 2. Aaj ke views enter karen (Aap yahan 4.85 dalen gi)
aaj_ki_tarikh = date.today().strftime("%b %d")
aaj_ke_views = float(input(f"Enter views for {aaj_ki_tarikh}: "))

# 3. Naya data list mein shamil karen
new_entry = pd.DataFrame({'Date': [aaj_ki_tarikh], 'Views': [aaj_ke_views]})
df = pd.concat([df, new_entry], ignore_index=True)

# 4. Data ko save karen
df.to_csv(file_name, index=False)

# 5. Automatic Graph dikhayen
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Views'], marker='o', color='green', linewidth=2)
plt.title('My Portfolio: Growth Tracking Bot', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Views (in thousands)')
plt.grid(axis='y', linestyle='--')

# Har point par value likhen
for i, txt in enumerate(df['Views']):
    plt.annotate(txt, (df['Date'][i], df['Views'][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()

print(f"Shabash Ishrat! Aapka data '{file_name}' mein mehfooz ho gaya hai.")
```

```python
import matplotlib.pyplot as plt

# آپ کے ناولز کا ڈیٹا
novels = ['Mera Hisar (Eng)', 'Mera Hisar (Urdu)', 'Mera Mahi (Urdu)', 'Khushiyon Mahal']
views = [5370, 258, 83, 41] # آپ کے تازہ ترین ویوز

# گراف کا ڈیزائن
plt.figure(figsize=(10, 6))
bars = plt.bar(novels, views, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# گراف پر نمبرز لکھنا
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, yval, ha='center', va='bottom', fontsize=12, fontweight='bold')

# ٹائٹل اور لیبلز
plt.title('Ishrat Zahid Novels: Views Analysis (Feb 2026)', fontsize=16, fontweight='bold')
plt.xlabel('Novel Titles', fontsize=12)
plt.ylabel('Number of Views', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# گراف دکھانا
plt.show()
```
